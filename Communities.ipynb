```python
from handlers.graphbuilder import GraphBuilder
from handlers.graphviewer import GraphViewer
```

```python
gb = GraphBuilder()
gb.createGraph('modernist-journals/Journal-edges.csv')
G = gb.getGraph()
gv = GraphViewer(G)
```

```python
#COLLABORAZIONI NELLO STESSO GIORNALE
gv.setLayout('kamada_kawai')
gv.drawGraph(edgesPartitioned=True)
gv.showLayout()
```

```python
#AUTORI CHE HANNO SCRITTO SOLO IN UN GIORNALE
gv.setLayout('kamada_kawai')
gv.drawGraph(nodesPartitioned=True)
gv.showLayout()
```

```python
#AUTORI CHE HANNO SCRITTO IN PIU GIORNALI
gv.setLayout('kamada_kawai')
gv.drawGraph(commonNodes=True)
gv.showLayout()
```

```python
#MAX CLIQUE
clique = gb.findMaxClique()
gv.setLayout('kamada_kawai')
gv.drawClique(clique)
```

```python
#TRIADI
gb.filterEdges(10)
gb.filterNodes(1)
triads = gb.findTriads(77)
gv.drawTableTriads(triads)
```

```python
from handlers.graphbuilder import GraphBuilder
from handlers.graphviewer import GraphViewer
```

```python
gb = GraphBuilder()
gb.createGraph('modernist-journals/Journal-edges.csv')
G = gb.getGraph()
gv = GraphViewer(G)
```

```python
#DEGREE CENTRALITY 
dictDegCentr = gb.analyzeCentrality('degree')
gv.setLayout("kamada_kawai")
gv.drawTableCentrality(dictDegCentr, 'Degree Centrality', 0.75)
gv.drawCentrality(dictDegCentr.values(), 200)
gv.showLayout()
gv.drawDistribution(dictDegCentr)
```

```python
#BETWEENNES CENTRALITY
dictBetCentr = gb.analyzeCentrality('betweenness')
gv.setLayout("spring")
gv.drawTableCentrality(dictBetCentr, 'Degree Centrality', 0.75)
gv.drawCentrality(dictBetCentr.values(), 200)
gv.showLayout()
gv.drawDistribution(dictBetCentr)
```

```python
#BETWEENNES CENTRALITY
dictCloCentr = gb.analyzeCentrality('closeness')
gv.setLayout("kamada_kawai")
gv.drawTableCentrality(dictCloCentr, 'Closeness Centrality', 0.9)
gv.drawCentrality(dictCloCentr.values(), 100)
gv.showLayout()
gv.drawDistribution(dictCloCentr)
```

```python
#BETWEENNES CENTRALITY
dictEigCentr = gb.analyzeCentrality('eigenvector')
gv.setLayout("kamada_kawai")
gv.drawTableCentrality(dictEigCentr, 'Eigenvector Centrality', 0.75)
gv.drawCentrality(dictEigCentr.values(), 200)
gv.showLayout()
gv.drawDistribution(dictEigCentr)
```
